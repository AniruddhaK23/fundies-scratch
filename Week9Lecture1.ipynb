```python
def product_list(nums):
  if len(nums) == 0:
    return 1
  else:
    product = 1
    for num in nums:
      product = product*num
    return product
```

```python
def count_occurances(items,target):
  counter = 0
  for ele in items:
    if ele == target:
      counter += 1
  return counter
```

```python
def filter_by_prefix(words,prefix):
  temp_L =[]
  for word in words:
    if word.startswith(prefix):
      temp_L.append(word)
  return temp_L

```

```python
L=[]
for key in sessions :
  if 'AI' in sessions[key]['topics']:
    L.append(key)
print(L)
```

```python
def uni_topics():
  temp_L=[]
  for key in sessions:
    for topic in sessions[key]['topics']:
      if topic not in temp_L:
        temp_L.append(topic)
  return temp_L
```
